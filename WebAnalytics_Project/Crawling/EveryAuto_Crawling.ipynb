```python
from bs4 import BeautifulSoup
import pandas as pd
import urllib.request
```

```python
num_of_reviews = 31000
count = 0
page = 0

date = []
review = []
rating = []

while num_of_reviews > count:
    
    error = True
    while error:
        try:
            page +=1
            url = 'https://everyauto.com/cars/research-reviews/?make=Toyota&a=1m1o1k1g1j1n1k1j1p&year=&page={}'.format(page)
            ourUrl = urllib.request.urlopen(url)
            soup = BeautifulSoup(ourUrl,'html.parser')
            error = False
        except Exception:
            error = True

    for s1 in soup.find_all('li',{'class':'user-review'}):
        #rating
        rating.append(float(s1.find('meta',{'itemprop':'ratingValue'}).get('content')))

        #review
        review.append(s1.find('p',{'class':'user-review-text'}).text[45:])

        #date
        date.append(s1.find('span',{'class':'post-date'}).get('content')[:10])

        count +=1
```

```python
df_Toyota = pd.DataFrame(list(zip(date,rating,review)),columns = ['Date','Rating','ReviewText'])
df_Toyota.to_csv("EveryAutoToyota2.csv")
df_Toyota
```
