```python
from selenium import webdriver
from bs4 import BeautifulSoup
import urllib.request
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By


import datetime
import time
import pandas as pd
import numpy as np
from datetime import date
```

```python
dates = []
review = []
rating = []

review_counts = 5820
review_counted = 0
page = 1

driver = webdriver.Chrome(executable_path='C:\\Users\\baris.baturay\\Desktop\\Fordham\\Web Analytics\\Project\\chromedriver')

url = 'https://www.glassdoor.com/Reviews/Ford-Motor-Company-Reviews-E263_P{}.htm?filter.iso3Language=eng'.format(page)

driver.get(url)

sign_in_path = '/html/body/header/nav[1]/div/div/div/div[1]/button'

driver.find_element(By.XPATH, sign_in_path).click()

email = driver.find_element(By.ID,'userEmail')
password = driver.find_element(By.ID,'userPassword')
email.send_keys("bbaturay97baris@gmail.com")
password.send_keys("baris97Bb.")

sign_in_path2 = "/html/body/div[11]/div/div/div[2]/div[2]/div[2]/div/div/div/div[3]/form/div[3]/div[1]/button"

driver.find_element(By.XPATH, sign_in_path2).click()

while review_counts > review_counted: 

    soup = BeautifulSoup(driver.page_source,'html.parser')

    for s1 in soup.find_all('li',{'class':'noBorder empReview cf pb-0 mb-0'}):

        if review_counts > review_counted:
            
            rating.append(float(s1.find('span',{'class':'ratingNumber mr-xsm'}).text))

            dates.append(s1.find('span',{'class':'authorJobTitle middle common__EiReviewDetailsStyle__newGrey'}).text.split("-")[0][:-1])

            text1 = s1.find('h2',{'class':'mb-xxsm mt-0 css-93svrw el6ke055'}).text
            text2 = s1.find('span',{'data-test':'pros'}).text.replace('\n','').replace('\r','').replace('-','') + s1.find('span',{'data-test':'cons'}).text.replace('\n','').replace('\r','').replace('-','')
            text3 = text1 + ' ' + text2

            review.append(text3)
            
            review_counted += 1
    
    # wait for 0.3 seconds to click the button
    time.sleep(3)
    
    nextpage_path = '/html/body/div[3]/div/div/div/div[1]/div[1]/div/div/article/div[7]/main/div[2]/div/div[1]/div[6]/div/div[1]/button[2]'
    
    driver.find_element(By.XPATH, nextpage_path).click()
    
    time.sleep(3)
    # update main page url
    url = driver.current_url

driver.close()   
```

```python
glassdoor_ford= pd.DataFrame(list(zip(dates, rating, review)),
                  columns = ['date', 'rating', 'review',])
glassdoor_ford
```
