```python
import pandas as pd 
import numpy as np
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
employee_toyota = pd.read_csv('C:\\Users\\sejal\\Documents\\Fall Courses\\Web Analytics\\Project\\Data\\employee_toyota.csv')
employee_toyota.dropna(how = 'any',inplace=True)
df = list(employee_toyota.review)
```

```python
dicti = []
for i in df:
    try:
        dicti.append(sia.polarity_scores(i))
    except:
        print(i)
```

```python
neu = []
comp = []
neg = []
pos = []

for i in dicti:
    neg.append (i.get('neg'))
    pos.append (i.get('pos'))
    neu.append (i.get('neu'))
    comp.append (i.get('compound'))
```

```python
employee_toyota['positive'] = pos
employee_toyota['negative'] = neg
employee_toyota['compound'] = comp
employee_toyota['neutral'] = neu
```
