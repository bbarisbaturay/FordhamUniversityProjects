```python
import pandas as pd 
import numpy as np
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
employee_ford = pd.read_csv('C:\\Users\\sejal\\Documents\\Fall Courses\\Web Analytics\\Project\\Data\\employee_ford.csv')
employee_ford.dropna(how = 'any',inplace=True)
df = list(employee_ford.review)
```

```python
dicti = []
for i in df:
    try:
        dicti.append(sia.polarity_scores(i))
    except:
        print(i)
```

```python
neu = []
comp = []
neg = []
pos = []

for i in dicti:
    neg.append (i.get('neg'))
    pos.append (i.get('pos'))
    neu.append (i.get('neu'))
    comp.append (i.get('compound'))
```

```python
employee_ford['positive'] = pos
employee_ford['negative'] = neg
employee_ford['compound'] = comp
employee_ford['neutral'] = neu
```
