```python
import pandas as pd 
import numpy as np
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
consumer_toyota = pd.read_csv('C:\\Users\\sejal\\Documents\\Fall Courses\\Web Analytics\\Project\\Data\\consumer_toyota.csv')
consumer_toyota.dropna(how = 'any',inplace=True)
df = list(consumer_toyota.review)
```

```python
dicti = []
for i in df:
    try:
        dicti.append(sia.polarity_scores(i))
    except:
        print(i)
```

```python
neu = []
comp = []
neg = []
pos = []

for i in dicti:
    neg.append (i.get('neg'))
    pos.append (i.get('pos'))
    neu.append (i.get('neu'))
    comp.append (i.get('compound'))
```

```python
consumer_toyota['positive'] = pos
consumer_toyota['negative'] = neg
consumer_toyota['compound'] = comp
consumer_toyota['neutral'] = neu
```
